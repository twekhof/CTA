**CTA-Tutorial: 2008 American National Election Survey (ANES)**

The ANES data contain a text response in which respondents describe their most important political problem. The data also includes the respondents' Republican or Democrat party affiliation. The results were used in CTA's academic paper to validate the method (Wekhof, 2024).

To replicate Wekhof (2024), use the following package versions:<br/>
-torch: 2.4.0<br/>
-transformers: 4.32.1

The dataset for the following example was also used by Roberts et al. (2014) to validate the Structural Topic Model (STM), and it is available on Harvard Dataverse. It is called "final_anes_metadata.csv" and is available at https://doi.org/10.7910/DVN/29405.

<br/><br/>
*References*

Roberts, M. E., Stewart, B. M., Tingley, D., Lucas, C., Leder-Luis, J., Gadarian, S. K., Albertson, B. and Rand, D. G.: 2014, Structural topic models for open-ended survey responses, American Journal of Political Science 58(4), 1064â€“1082.

Wekhof, Tobias. "Conditional Topic Allocations for Open-Ended Survey Responses." Available at SSRN 4190308 (2024).

```python
import CTApy
import pandas as pd
import numpy as np
import random
import os

import statsmodels.api as sm
import matplotlib.pyplot as plt
import numpy as np
```

```python
#define paths
path_data='' # path to data
path_graphics=''
path_fasttext='' # path to fasttext embeddings
name_fasttext='cc.en.300.bin' # this is the English fasttext model, for other languages, use the respective model
spacy_model='en_core_web_sm' # this is the English spacy model, for other languages, use the respective model
```

```python
# load data
df = pd.read_csv(path_data+"final_anes_metadata.csv",sep=",",dtype=str)
```

```python
# define variables

# remove rows with ideology (-4)
# identify text and ourcome
df['ideology']=pd.to_numeric(df['ideology'])
df['pid_summary']=pd.to_numeric(df['pid_summary'])
df=df[df['pid_summary']>0]
df=df[df['pid_summary']!=3].reset_index(drop=True)
df['republican']=[1 if x>3 else 0 for x in df['pid_summary']]

text_var='mip_1'
outcome_var='republican'
metadata_vars=['miiele1_code1','mippol1_code1','age','female']
fasttext_model=path_fasttext+name_fasttext
```

```python
# preprocess data: this will give a df with preprocessed data and the embedding model

data_processed,embedding_model=CTApy.preprocessing(df, text_var,outcome_var,metadata_vars,spacy_model,fasttext_model)
```

```python
#save preprocessed data to run other topic models
#data_processed.to_csv(path_data+"cta_data/anes_lemmatized.csv", index=False)
```

```python
# parameters for prediction step with DistilBERT
bert_model='distilbert-base-uncased' # this is the English DistilBERT model, for other languages, use the respective model
seed=42
epochs=20
```

```python
# prediction step: it returns the trained distilBERT model and the tokenizer

model, token = CTApy.prediction(data_processed['text_lemma'],data_processed['outcome'],seed,bert_model,epochs)
```

```python
# compute word importance: it returns a matrix with all SHAP values; min_len and min_freq is the minimum number of
# characters and word occurence for the words considered for the SHAP values matrix

shap_words=CTApy.word_importance(model, token, data_processed['text_lemma'],min_len=3,min_freq=2)
```

```python
#name for the shap value file, e.g., anes_republican_shap_bert_180824.csv

#shap_words.to_csv(path_data+"cta_data/anes_republican_shap_bert_180824.csv", index=False)
```

```python
# load SHAP words matrix if necessary

#shap_words=pd.read_csv(path_data+"cta_data/anes_republican_shap_bert_180824.csv")
```

```python
# set parameters for topic allocation (step 3)

p_val=0.1 # p-value cutoff to sort words into positive or negative impact group
method='top_shap' # type of ordering of topics; 
# available options: 'top_shap', 'mean_shap', 'impact', 'freq_only', and 'cohecence'

n_topics=2 # number of topics that will be shown, according to the ordering from "method"
```

```python
# topic allocation (step 3): it returns a list with the topics positively associated with the outcome and negatively associated
# the inputs are the SHAP matrix, the cutoff p-value, ordering method, number of topics to show, preprocessed data, and
# the embedding model 

topics_pos, topics_neg=CTApy.allocate_topics(shap_words,p_val,method,n_topics,data_processed['text_lemma'],embedding_model)
```

```python
# label the pos/neg topics and tag each individual text answer with the respoective topic

name_pos='Republican'
name_neg='Democrat'
text=data_processed['text_lemma']
outcome=data_processed['outcome']

# tags the text answers with CTA topics; multiple topics per response possible
# inputs are the text, the pos and neg topic lists, the SHAP matrix, the topic names, and the tag type
# for the tag type there are two possibilities: 'shap' (using shap values) or 'indicator', which allocates a dummy variable for
# each topic

df_reg=CTApy.tag_text(text,outcome,topics_pos,topics_neg,shap_words,name_pos,name_neg,tag_type='shap')
```

```python
#####################
# Descriptive Plots #
#####################

# Convert DataFrame to numeric data type
df_reg = df_reg.apply(pd.to_numeric, errors='coerce')

# Store coefficients and confidence intervals
coefficients = []
conf_intervals = []

# Run a linear regression for each topic dummy
for column in df_reg.columns:
    if column != 'outcome':
        X = df_reg['outcome']
        y = df_reg[column]
        X = sm.add_constant(X)  # adding a constant

        model = sm.OLS(y, X).fit()
        coefficients.append(model.params.iloc[1])  # Use iloc to access by position
        conf_intervals.append(model.conf_int(alpha=0.1).iloc[1].values)


# Plot coefficients with confidence intervals
# Define your desired font size and marker size
font_size = 18
marker_size = 10
line_width = 2

# Plot coefficients with confidence intervals
plt.figure(figsize=(10, 5))

topic_order=np.arange(len(coefficients),0,-1)-1
# Plot error bars with increased marker size and line width
plt.errorbar(coefficients, topic_order, 
             xerr=[abs((top-bot)/2) for top, bot in conf_intervals],
             fmt='o', markersize=marker_size, elinewidth=line_width)

# Increase font size for y-axis labels
plt.yticks(topic_order, df_reg.columns[df_reg.columns != 'outcome'], fontsize=font_size)

# Get the current limits of the x-axis
xmin, xmax = plt.xlim()

# Define x-axis labels and their positions
labels = ['Democrat', u'\u2190 Correlation \u2192', 'Republican']  # Unicode characters for left and right arrows
positions = [xmin, (xmin + xmax) / 2, xmax]  # Place 'Correlation' at the middle of the x-axis

# Add new x-axis labels
for pos, label in zip(positions, labels):
    plt.text(pos, -0.15, label, fontsize=font_size, ha='center', va='top', transform=plt.gca().get_xaxis_transform())

# Increase font size for original x-axis labels
plt.xticks(fontsize=font_size)

plt.axvline(x=0, color='black', linewidth=1)

#plt.savefig(path_graphs+'cta_topic_outcome_anes.png',bbox_inches='tight')

plt.show()
```

```python
# Plot embeddings for each step

# create word embeddings for clustering
from gensim.models import fasttext
# Load pre-trained model (for another language, use a different pre-trained model)
pretrained_model = fasttext.load_facebook_model("research_data/cc.en.300.bin")
```

```python

pretrained_model.build_vocab(corpus_iterable=data_processed['text_lemma'], update=True)
pretrained_model.train(corpus_iterable=data_processed['text_lemma'], 
                       total_examples=len(data_processed['text_lemma']), epochs=pretrained_model.epochs)
```

```python
#all words
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE

# Set seed for reproducibility
np.random.seed(42)

# Get the words from feature_importance['Variable']
important_features = shap_words.columns.tolist()

# Filter feature_names
filtered_feature_names = [word for word in important_features if word in pretrained_model.wv.key_to_index]

# Get the word vectors for the words in filtered_feature_names
word_vectors = [pretrained_model.wv[word] for word in filtered_feature_names]

# Convert list to numpy array
word_vectors_np = np.array(word_vectors)

# Use t-SNE for dimensionality reduction
tsne_model = TSNE(perplexity=40, n_components=2, init='pca', max_iter=2500, random_state=42)
new_values = tsne_model.fit_transform(word_vectors_np)

# Create a dataframe
tsne_df_step1 = pd.DataFrame({'X':new_values[:, 0], 'Y':new_values[:, 1], 'Word':filtered_feature_names})

# Plot the word embeddings
plt.figure(figsize=(5, 5)) 
for i in range(len(tsne_df_step1)):
    plt.scatter(tsne_df_step1.X[i],tsne_df_step1.Y[i], color='black', s=30)  # s parameter controls the size of the dots
plt.show()
```

```python
shap_values_matrix=shap_words
from scipy.stats import sem
from scipy import stats
feature_importance = pd.DataFrame({
    'Variable': shap_values_matrix.columns,
    'word_importance': shap_values_matrix.mean(),
    'word_se': sem(shap_values_matrix),
    'word_tval': [stats.ttest_1samp(shap_values_matrix[col], 0)[0] for col in shap_values_matrix.columns],
    'word_pval': [stats.ttest_1samp(shap_values_matrix[col], 0)[1] for col in shap_values_matrix.columns]
    }).reset_index(drop=True)
feature_importance = feature_importance.fillna(0)
feature_importance_pos=feature_importance[(feature_importance['word_pval']<p_val)&
                                      (feature_importance['word_importance']>0)].reset_index(drop=True)
feature_importance_neg=feature_importance[(feature_importance['word_pval']<p_val)&
                                      (feature_importance['word_importance']<0)].reset_index(drop=True)
```

```python
#positive words

# Set seed for reproducibility
np.random.seed(42)

# Get the words from feature_importance['Variable']
important_features = feature_importance_neg['Variable'].tolist()

# Filter feature_names
filtered_feature_names = [word for word in important_features if word in pretrained_model.wv.key_to_index]

# Get the word vectors for the words in filtered_feature_names
word_vectors = [pretrained_model.wv[word] for word in filtered_feature_names]

# Convert list to numpy array
word_vectors_np = np.array(word_vectors)

# Use t-SNE for dimensionality reduction
tsne_model = TSNE(perplexity=20, n_components=2, init='pca', max_iter=2500, random_state=42)
new_values = tsne_model.fit_transform(word_vectors_np)

# Create a dataframe
tsne_df_step2 = pd.DataFrame({'X':new_values[:, 0], 'Y':new_values[:, 1], 'Word':filtered_feature_names})

# Plot the word embeddings
plt.figure(figsize=(5, 5)) 
for i in range(len(tsne_df_step2)):
    plt.scatter(tsne_df_step2.X[i],tsne_df_step2.Y[i], color='black', s=30)  # s parameter controls the size of the dots
plt.show()
```

```python
# plot positive topics

from sklearn.cluster import KMeans
from matplotlib.patches import Circle
from matplotlib.collections import PatchCollection
from scipy.spatial import ConvexHull
import matplotlib.patches as mpatches


# Assign colors to each group
group_colors = ['red', 'blue', 'green', 'orange']

# Filter word vectors and words based on the subset
subset_word_vectors = []
subset_words = []
for group in topics_neg:
    for word in group:
        if word in pretrained_model.wv.key_to_index:
            subset_word_vectors.append(pretrained_model.wv[word])
            subset_words.append(word)

# Convert list to numpy array
subset_word_vectors_np = np.array(subset_word_vectors)

# Use t-SNE for dimensionality reduction
tsne_model = TSNE(perplexity=8, n_components=2, init='pca', max_iter=2500, random_state=42)
new_values = tsne_model.fit_transform(subset_word_vectors_np)

# Create a dataframe
tsne_df_step3 = pd.DataFrame({'X':new_values[:, 0], 'Y':new_values[:, 1], 'Word':subset_words})

# Plot the word embeddings
fig, ax = plt.subplots(figsize=(6, 6))

# Add encircling for each group with colors
for i, group in enumerate(topics_neg):
    group_vectors = [new_values[j] for j in range(len(new_values)) if subset_words[j] in group]
    if len(group_vectors) > 2:  # ConvexHull requires at least 3 points
        hull = ConvexHull(group_vectors)
        polygon_points = [group_vectors[vertex] for vertex in hull.vertices]
        poly = mpatches.Polygon(polygon_points, closed=True, alpha=0.3, color=group_colors[i])
        ax.add_patch(poly)

# Create the base scatter plot
plt.scatter(tsne_df_step3.X, tsne_df_step3.Y, color='black', s=10)
plt.show()
```
